```python
import requests 
from bs4 import BeautifulSoup
import json
```

```python
url = "https://remoteok.io/" #downloads the HTML content of the page.
headers = {"User-Agent": "Mozilla/5.0"} #makes your script look like a normal web browser (some websites block â€œbotâ€ requests if you donâ€™t include this).
response = requests.get(url, headers = headers)#contains the HTML source code of the page.
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup)

#BeautifulSoup parses the HTML text into a structured format (like a tree).
#Now we can search for tags like <tr>, <h2>, <h3>, etc.
```

```python
soup.contents # for displaying
```

```python
print(soup.prettify()) # for prettifying the output
```

```python
jobs =[]
job_cards = soup.find_all("tr", class_="job")

#On RemoteOK, each job listing is inside a <tr class="job">...</tr>.

#find_all("tr", class_="job") â†’ gets all job rows.
```

```python
#Extract Data From Each Job
for job in job_cards:
    title = job.find("h2",itemprop="title")
    company = job.find("h3",itemprop="name")
    tags = job.find_all("span", class_="tag")
    date = job.find("time")
    link = job.get("data-href")
    
#Store Data in a Dictionary
    job_info = {
        "title" : title.text.strip() if title else None,
        "company" : company.text.strip() if company else None,
        "tags" : [tag.text.strip() for tag in tags] if tags else [],
        "date_posted": date["datetime"] if date else None,
        "url" : f"https://remoteok.io{link}" if link else None
    }
    jobs.append(job_info)

```

```python
#Save Data into JSON File
with open("jobs.json", "w", encoding="utf-8") as f:
    json.dump(jobs , f , ensure_ascii= False , indent = 4)

print("âœ… Scraping complete! Jobs saved to jobs.json")
```

âš¡ So the workflow is basically:
Request page â†’ Parse HTML â†’ Locate elements â†’ Extract â†’ Store in JSON.
